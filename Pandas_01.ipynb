#### Removed Rows (rows had cells with empty values).

```python
df.dropna(inplace = True)
print(df.to_string())
```

## Replace Empty values

```python
df.fillna(130,inplace=True)
```

```python
#spacific column
df["Pulse"].fillna(130, inplace=True)
```

# means

```python
x=df["Calories"].mean()
y=df["Pulse"].mean()
n=df["Maxpulse"].mean()
m=df["Duration"].mean()
```

```python
print("Mean of Calories",x)
print("Mean of Pulse",y)
print("Mean of Maxpulse",n)
print("Mean of Duration",m)
```

# medians

```python
a=df["Calories"].median()
b=df["Pulse"].median()
c=df["Maxpulse"].median()
d=df["Duration"].median()
```

```python
print("Median of Calories",a)
print("Median of Pulse",b)
print("Median of Maxpulse",c)
print("Median of Duration",d)
```

# Adding new Date coloumn in my dataset

```python
df['Date'] = pd.date_range(start='2023-01-01', periods=len(df), freq='D')

print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.to_string())
```

## Replace valuse

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.loc[x,"Duration"]=120
    else:
        df.loc[x,"Duration"]=0
print(df.to_string())
```

## finding a duplicate 

```python
print(df.duplicated())
```

```python
#Remove all Duplicates
df.drop_duplicates(inplace=True)
print(df.to_string())
```
