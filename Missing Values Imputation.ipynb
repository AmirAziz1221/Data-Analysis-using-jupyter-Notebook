# Advanced Data Cleaning Techniques

### Step 1: Handling Categorical Variables
  - Encoding Categorical Variables

```python
# Convert categorical variables into numerical representations (e.g., 'sex', 'embarked')
titanic = pd.get_dummies(titanic, columns=['sex'], drop_first=True) #'Embarked' is also if present
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode 'Sex_male' column
titanic['sex_encoded'] = label_encoder.fit_transform(titanic['sex_male'])
```

```python
titanic.head()
```

### Step 2: Dealing with Textual Data
- Feature Extraction from 'Name' Column

```python
# Extract titles from the 'name' column
titanic['title'] = titanic['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
# Check unique titles and potentially group them
print(titanic['title'].value_counts())

# Map rare titles to a common category (e.g., 'Rare')
title_mapping = {'Capt': 'Rare', 'Col': 'Rare', 'Don': 'Rare', 'Dr': 'Rare', 'Major': 'Rare',
                 'Rev': 'Rare', 'Jonkheer': 'Rare', 'Dona': 'Rare', 'Lady': 'Rare',
                 'Countess': 'Rare', 'Sir': 'Rare'}
titanic['title'] = titanic['title'].replace(title_mapping)

# Convert titles to numerical values using LabelEncoder or get_dummies
```

### Step 3: Handling Age or Fare Binning
- Binning Numerical Variables (Age, Fare)

```python
# Binning 'age' into categories
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 80], labels=['child', 'young_adult', 'adult', 'elderly'])

# Binning 'fare' into categories
titanic['fare_category'] = pd.qcut(titanic['fare'], q=4, labels=['low', 'medium', 'high', 'very_high'])
```
