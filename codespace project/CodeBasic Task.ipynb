### Top 10 Batsmen Based on Total Runs Scored in the Past 3 Years

```python
# Grouping data by batsmanName and summing up the runs scored
batsmen_total_runs = df3.groupby('batsmanName')['runs'].sum()

# Sorting the batsmen based on total runs scored in descending order
top_10_batsmen = batsmen_total_runs.sort_values(ascending=False).head(10)

# Plotting the top 10 batsmen based on total runs scored
plt.figure(figsize=(10, 6))
top_10_batsmen.plot(kind='bar', color='skyblue')
plt.title('Top 10 Batsmen Based on Total Runs Scored (Past 3 Years)')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs Scored')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Top 10 Batsmen Based on Batting Average in the Past 3 Years (Minimum 60 Balls Faced in Each Season)

```python
# Convert 'SR' column to string type
df3['SR'] = df3['SR'].astype(str)

# Remove '%' symbol and convert 'SR' column to numeric
df3['SR'] = pd.to_numeric(df3['SR'].str.rstrip('%'), errors='coerce')

# Filtering batsmen who have faced at least 60 balls in each season
batsmen_60_balls = df3.groupby('batsmanName').filter(lambda x: (x['balls'] >= 60).all())

# Calculating batting average for each batsman
batsmen_60_balls['batting_average'] = batsmen_60_balls['runs'] / batsmen_60_balls['out/not_out'].apply(lambda x: 1 if x == 'out' else 0)

# Grouping data by batsmanName and calculating the mean batting average
batsmen_avg = batsmen_60_balls.groupby('batsmanName')['batting_average'].mean()

# Sorting the batsmen based on batting average in descending order
top_10_avg_batsmen = batsmen_avg.sort_values(ascending=False).head(10)

print(top_10_avg_batsmen)

# Check if there are any batsmen who have faced at least 60 balls in each season
if not batsmen_60_balls.empty:
    # Plotting the top 10 batsmen with the highest batting average
    plt.figure(figsize=(10, 6))
    top_10_avg_batsmen.plot(kind='bar', color='skyblue')
    plt.title('Top 10 Batsmen by Batting Average')
    plt.xlabel('Batsman')
    plt.ylabel('Batting Average')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No data available for visualization. There are no batsmen who have faced at least 60 balls in each season.")
```

### Top 10 Bowlers Based on Bowling Average in the Past 3 Years (Minimum 60 Balls Bowled in Each Season)

```python
# Filtering bowlers who have bowled at least 60 balls in each season
bowlers_60_balls = df4.groupby('bowlerName').filter(lambda x: (x['overs'] * 6 >= 60).all())

# Calculating bowling average for each bowler
bowlers_60_balls['bowling_average'] = bowlers_60_balls['runs'] / bowlers_60_balls['wickets']

# Grouping data by bowlerName and calculating the mean bowling average
bowlers_avg = bowlers_60_balls.groupby('bowlerName')['bowling_average'].mean()

# Sorting the bowlers based on bowling average in ascending order
top_10_avg_bowlers = bowlers_avg.sort_values().head(10)

print(top_10_avg_bowlers)

# Assuming top_10_avg_bowlers is already calculated

if not top_10_avg_bowlers.empty:
    # Plotting the top 10 bowlers with the lowest bowling average
    plt.figure(figsize=(10, 6))
    top_10_avg_bowlers.plot(kind='bar', color='lightblue')
    plt.title('Top 10 Bowlers by Bowling Average')
    plt.xlabel('Bowler')
    plt.ylabel('Bowling Average')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("There are no bowlers who have bowled at least 60 balls in each season.")
```

### Top 10 Bowlers Based on Economy Rate in the Past 3 Years (Minimum 60 Balls Bowled in Each Season)

```python
# Filtering bowlers who have bowled at least 60 balls in each season
bowlers_60_balls = df4.groupby('bowlerName').filter(lambda x: (x['overs'] * 6 >= 60).all())

# Calculating economy rate for each bowler
bowlers_60_balls['economy_rate'] = bowlers_60_balls['runs'] / bowlers_60_balls['overs']

# Grouping data by bowlerName and calculating the mean economy rate
bowlers_avg_economy = bowlers_60_balls.groupby('bowlerName')['economy_rate'].mean()

# Sorting the bowlers based on economy rate in ascending order
top_10_economy_bowlers = bowlers_avg_economy.sort_values().head(10)

print(top_10_economy_bowlers)

# Assuming top_10_economy_bowlers is already calculated

if not top_10_economy_bowlers.empty:
    # Plotting the top 10 bowlers with the lowest average economy rate
    plt.figure(figsize=(10, 6))
    top_10_economy_bowlers.plot(kind='bar', color='lightgreen')
    plt.title('Top 10 Bowlers by Average Economy Rate')
    plt.xlabel('Bowler')
    plt.ylabel('Average Economy Rate')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("No bowlers meet the criteria for minimum balls bowled.")
```

### Top 5 Bowlers Based on Dot Ball Percentage in the Past 3 Years

```python
import numpy as np

# Assuming df4 contains the dataframe with bowling statistics

# Calculate total balls bowled by each bowler
df4['total_balls'] = df4['maiden'] + df4['wides'] + df4['noBalls']

# Replace 0 values with NaN in total balls to avoid division by zero
df4['total_balls'] = df4['total_balls'].replace(0, np.nan)

# Calculate dot ball percentage for each bowler
df4['dot_ball_percentage'] = (df4['0s'] / df4['total_balls']) * 100

# Fill NaN values with 0 in dot ball percentage
df4['dot_ball_percentage'] = df4['dot_ball_percentage'].fillna(0)

# Group by bowlerName and calculate the mean dot ball percentage
top_bowlers = df4.groupby('bowlerName')['dot_ball_percentage'].mean().reset_index()

# Sort the dataframe by dot ball percentage in descending order
top_bowlers = top_bowlers.sort_values(by='dot_ball_percentage', ascending=False)

# Get the top 5 bowlers
top_5_bowlers = top_bowlers.head(5)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_5_bowlers['bowlerName'], top_5_bowlers['dot_ball_percentage'], color='skyblue')
plt.title('Top 5 Bowlers Based on Dot Ball Percentage')
plt.xlabel('Bowler')
plt.ylabel('Dot Ball Percentage')
plt.xticks(rotation=45)
plt.yticks(np.arange(0, max(top_5_bowlers['dot_ball_percentage']) + 10, 10))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

### Orange and Purple Cap Players:

```python
# Calculate total runs scored by each batsman
total_runs = df3.groupby('batsmanName')['runs'].sum().reset_index()

# Calculate total wickets taken by each bowler
total_wickets = df4.groupby('bowlerName')['wickets'].sum().reset_index()

# Identify the Orange Cap player (most runs)
orange_cap_player = total_runs.loc[total_runs['runs'].idxmax()]

# Identify the Purple Cap player (most wickets)
purple_cap_player = total_wickets.loc[total_wickets['wickets'].idxmax()]

print("Orange Cap Player:", orange_cap_player['batsmanName'])
print("Purple Cap Player:", purple_cap_player['bowlerName'])

# Visualization for Orange and Purple Cap Players
plt.figure(figsize=(10, 5))
plt.bar(['Orange Cap', 'Purple Cap'], [orange_cap_player['runs'], purple_cap_player['wickets']], color=['orange', 'purple'])
plt.title('Orange and Purple Cap Players for 2024 Season')
plt.ylabel('Count')
plt.xlabel('Player Type')
plt.show()
```

### Winner and Runner-up

```python
# Calculate total wins for each team
team_wins = df1.groupby('winner')['winner'].count().reset_index(name='wins')

# Identify the winner (team with most wins)
winner = team_wins.loc[team_wins['wins'].idxmax()]

# Remove the winner from the dataframe to find the runner-up
team_wins = team_wins[team_wins['winner'] != winner['winner']]

# Identify the runner-up (team with second most wins)
runner_up = team_wins.loc[team_wins['wins'].idxmax()]

print("Winner:", winner['winner'])
print("Runner-up:", runner_up['winner'])

# Visualization for Winner and Runner-up
plt.figure(figsize=(8, 5))
sns.barplot(x='wins', y='winner', data=team_wins, palette='coolwarm')
plt.title('Winner and Runner-up for 2024 Season')
plt.xlabel('Wins')
plt.ylabel('Team')
plt.show()
```
